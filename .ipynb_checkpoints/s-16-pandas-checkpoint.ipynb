```python
import numpy as np
import pandas as pd
```

###serise from list

```python
country = ["india", "usa", "china", "japan", "germany"]
pd.Series(country)
```

```python
#integers
runs = [100, 200, 300, 400, 500]
pd.Series(runs)
```

```python
#custom index
marks = [100, 200, 300, 400, 500]
subjects = ['math','phy','chem','bio','english']
pd.Series(marks, index=subjects)
```

```python
#setting a name
marks = [100, 200, 300, 400, 500]
subjects = ['math','phy','chem','bio','english']
aji_marks = pd.Series(marks, index=subjects,name='ajinkya ke marks')
aji_marks
```

serise from dict

```python
marks = {
    'math':100,
    'phy':200,
    'chem':300,
    'bio':400,
    'english':500
}

aj_marks = pd.Series(marks,name = "ajinkya ke marks")
aj_marks
```

series attributes

```python
#size (tells values num)
aj_marks.size
```

```python
#dtype (tells data type of values)
aj_marks.dtype
```

```python
#names (tells name of series)
aji_marks.name
```

```python
#is_unique (tells if values are different or not)
aj_marks.is_unique
```

```python
#index (tells index values)
aj_marks.index
```

```python
#values (tells values in series)
aj_marks.values
```

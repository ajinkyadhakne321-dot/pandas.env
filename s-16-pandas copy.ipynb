#sclicing

```python
# Get entire column as Series
ajdata['ProductRelated']

# Slice first 5 values from a column
ajdata['ProductRelated'][0:5]
ajdata['ProductRelated'].iloc[0:5]

# Slice with step
ajdata['ProductRelated'][::2]  # Every 2nd value

# Using loc (label-based)
ajdata.loc[0:5, 'ProductRelated']

# Using iloc (position-based)
ajdata.iloc[0:5, 0]  # First 5 rows of 1st column

# Multiple columns
ajdata[['ProductRelated', 'Administrative']][0:5]

# Slice with boolean condition
ajdata['ProductRelated'][ajdata['ProductRelated'] > 10]

# Get specific positions
ajdata['ProductRelated'][[0, 2, 4]]  # Values at positions 0, 2, 4
```

indexing with labels

```python
# label-based indexing using .loc
# access a single value by row label and column name
ajdata.loc[0, 'ProductRelated']

# slice rows by label and select a column
ajdata.loc[0:5, 'ProductRelated']

# select multiple columns for specific row labels
ajdata.loc[0:3, ['ProductRelated', 'Administrative']]

# when index has custom labels, you can use them directly
# e.g. if ajdata.index = ['a','b','c',...], then ajdata.loc['a','ProductRelated']
```
